```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt 

women_degrees = pd.read_csv("C:/Users/Kristine/Documents/percent-bachelors-degrees-women-usa.csv")
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics'] 
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']

fig = plt.figure(figsize=(9,18))

for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]],c=cb_dark_blue, linewidth=2, label='Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[cat_index]], c=cb_orange, linewidth=2, label='Men')
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off', bottom='off', left='off', right='off', labelbottom='off')
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index], fontsize=10)
    if cat_index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')

for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, linewidth=2, label='Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[cat_index]], c=cb_orange, linewidth=2, label='Men')
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off', bottom='off', left='off', right='off', labelbottom='off')
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index], fontsize=10)
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
        ax.tick_params(labelbottom='on')

for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, linewidth=2, label='Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[cat_index]], c=cb_orange, linewidth=2, label='Men')
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off', bottom='off', left='off', right='off', labelbottom='off')
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index], fontsize=10)
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom='on')

plt.savefig('gender_degrees.png')
plt.show()
```
